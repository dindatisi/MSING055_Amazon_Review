## Most reviews made in a day
In this section we'll calculate the highest number of reviews that a profile made within a day. The logic is pretty straightforward since the date is already sorted from most recent:
- for every profile_id, check the date 
- if the current date row has the same value as the previous one, we'll add 1 to the counter `seq` 
- we compare `seq` for each day, and put the highest value to the dictionary `pid_seq`,with the `profile_id` as the key
