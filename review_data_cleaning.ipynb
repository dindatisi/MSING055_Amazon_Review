# DATA CLEANING CODE

## cleaning verified column

```python
import pandas as pd
import csv

profile_data = pd.read_csv("", quoting=csv.QUOTE_NONNUMERIC) #PUT FILE NAME HERE
profile_data
```

```python
verified_list = profile_data['verified'].tolist()
verified_list
```

```python
new_verified=[0 if x!="Verified Purchase" else 1 for x in verified_list]
```

```python
new_verified
```

```python
profile_data['verified'] = new_verified
```

```python
profile_data
```

## cleaning helpful column

```python
profile_data['helpful'] = profile_data['helpful'].fillna(0)
profile_data
```

```python
profile_data['product_link'] = profile_data['product_link'].fillna(0)
profile_data
```

## cleaning date column

```python
new_date=profile_data['date'].tolist()
new_date
```

```python
date_reformat=[(str(item))[3:] for item in new_date]
date_reformat
```

```python
profile_data['date'] = date_reformat
profile_data
```

## reformat stars

```python
stars=profile_data['ratings'].tolist()
new_stars=[int((str(item))[:-17]) for item in stars]
new_stars
```

```python
profile_data['ratings'] = new_stars
profile_data
```
